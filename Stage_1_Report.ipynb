## Task 2: Dataset Exploration
### Importing the Dataset
Here we download the dataset directly from kaggle.

```python
# Only need to run once to download dataset locally.
!pip install kagglehub pandas
import kagglehub
path = kagglehub.dataset_download("tmdb/tmdb-movie-metadata")
```

Next we load the 2 datasets into dataframs 'movies' and 'credits'.

```python
# Need to run every time to import dataset into notebook.
import os
import pandas as pd

# import movies

csv_file = os.path.join(path, "tmdb_5000_movies.csv")
movies = pd.read_csv(csv_file)

# import credits

csv_file = os.path.join(path, "tmdb_5000_credits.csv")
credits = pd.read_csv(csv_file)
```

We perform an outer merge on the movie id, combining the dataframes.

```python
# merge
df = pd.merge(movies, credits, left_on='id', right_on='movie_id', how='outer')
```

### Dataset Dictionary
This code creates a dataset dictionary listing our 25 features, their data type, an example entry, and % entries missing. We also assigned revenue as our target feature.

```python
# Task 2: Create a dataset dictionary
dataset_dict = pd.DataFrame({
    "Feature": df.columns,
    "Type": [df[col].dtype for col in df.columns],
    "Example": [df[col].dropna().iloc[0] for col in df.columns],  
    "Missing %": [df[col].isna().mean() * 100 for col in df.columns]
})

# Add target variable at the end
dataset_dict.loc[len(dataset_dict)] = [
    "Target Variable (revenue)",  
    df["revenue"].dtype,         
    df["revenue"].dropna().iloc[0], 
    df["revenue"].isna().mean() * 100 
]

dataset_dict
```

## Task 3 — Feature Exploration 
### Abdel
I cleaned a few features:

- runtime: filled missing values with the median
- budget and revenue: replaced 0 with NaN
- homepage and tagline: dropped because they had too many missing

Now the data looks cleaner for later analysis.

```python
import numpy as np

df["runtime"] = df["runtime"].fillna(df["runtime"].median())

df["budget"] = df["budget"].replace(0, np.nan)
df["revenue"] = df["revenue"].replace(0, np.nan)

df = df.drop(["homepage","tagline"], axis=1)

print("cleaned runtime, budget, revenue, homepage, tagline")
```

### Matthew

Before merging, both data sets had identical 'id' and 'title' columns, which created duplicates in the merge. So, I removed the duplicates, making a single 'id' and 'title' column.

```python
df = df.drop(["movie_id"], axis=1)
df = df.drop(["title_y"], axis=1)
df = df.rename(columns={"title_x": "title"})
```

Lots of the columns were hard to read, and looked to be stored as a dictionary, so I converted them into comma separated lists.  
Example: '[{"id": 80, "name": "Crime"}, {"id": 35, "name": "Comedy"}]' becomes 'Crime, Comedy'

```python
import ast

# List of columns that need conversion
json_cols = ['genres', 'keywords', 'cast', 'production_companies', 'production_countries', 'spoken_languages', 'crew']

for col in json_cols:
    df[col] = df[col].apply(lambda s: ", ".join(d['name'] for d in ast.literal_eval(s)) if s else "")
```

Reordered columns to group similar columns, and follow the general order of id, target, numeric data, lists, metadata.

```python
df = df[['id', 'revenue', 'budget', 'runtime', 'popularity', 'vote_average', 'vote_count', 'genres', 'keywords', 'cast', 'crew', 'production_companies', 'production_countries', 'spoken_languages', 'title', 'original_title', 'overview', 'release_date', 'status', 'original_language']]
```

```python
numeric_cols = ['budget', 'runtime', 'popularity', 'vote_average', 'vote_count']

# Compute mean and standard deviation
means = df[numeric_cols].mean().round()
stds = df[numeric_cols].std().round()

# Define outlier thresholds: mean ± 3*std
thresholds = pd.DataFrame({
    'Lower': (means - 3*stds).round(),
    'Upper': (means + 3*stds).round()
})

# Print outlier values, means, and thresholds
for col in numeric_cols:
    lower = int(thresholds.loc[col, 'Lower'])
    upper = int(thresholds.loc[col, 'Upper'])
    print(f"{col} mean: {int(means[col])}")
    print(f"{col} outlier thresholds: {lower} - {upper}")
    
    high_outliers = df[df[col] > upper][col].round().astype(int).values
    low_outliers = df[df[col] < lower][col].round().astype(int).values
    
    print(f"{col} high outliers: {high_outliers}")
    print(f"{col} low outliers: {low_outliers}\n")
```

```python
import numpy as np
from scipy.stats import zscore

numeric_cols = ['budget', 'runtime', 'popularity', 'vote_average', 'vote_count']

# Compute mean and std for outlier thresholds
means = df[numeric_cols].mean()
stds = df[numeric_cols].std()

# Define thresholds
lower = means - 3*stds
upper = means + 3*stds

# Identify outliers per column
outliers = (df[numeric_cols] < lower) | (df[numeric_cols] > upper)

# Identify NaNs per column
nans = df[numeric_cols].isna()

# Rows with any outlier
rows_with_outlier = outliers.any(axis=1)

# Rows with any NaN
rows_with_nan = nans.any(axis=1)

total_rows = len(df)
num_rows_with_outlier = rows_with_outlier.sum()
num_rows_with_nan = rows_with_nan.sum()

print(f"Total rows: {total_rows}")
print(f"Rows with any outlier: {num_rows_with_outlier}")
print(f"Rows with any NaN: {num_rows_with_nan}")
```

```python
numeric_cols = ['budget', 'runtime', 'popularity', 'vote_average', 'vote_count']

# Compute mean and std for outlier thresholds
means = df[numeric_cols].mean()
stds = df[numeric_cols].std()

# Define thresholds
lower = means - 3*stds
upper = means + 3*stds

# For each row, check numeric columns for outliers
for idx, row in df.iterrows():
    outlier_list = []
    for col in numeric_cols:
        if pd.isna(row[col]):
            continue  # skip NaN
        if row[col] > upper[col]:
            outlier_list.append(f"{col}: high")
        elif row[col] < lower[col]:
            outlier_list.append(f"{col}: low")
    
    if outlier_list:  # Only print rows with at least one outlier
        print(f"Title: {row['title']}")
        print(f"Outliers: {outlier_list}\n")
```
